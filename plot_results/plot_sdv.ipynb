```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


def read_data(filePath, prop):
    output_data = np.loadtxt(filePath, skiprows=2)
    columns=["X", prop]
    df = pd.DataFrame(output_data, columns=columns)
    prop_data = df.iloc[:, 1].to_numpy()[1:]
    return prop_data
```

```python
PEEQ = read_data("export_PEEQ.txt", "PEEQ")
strain_rate = read_data("export_strain_rate.txt", "strain_rate")
stress = read_data("export_stress.txt", "stress")
temperature = read_data("export_temperature.txt", "temperature")
```
